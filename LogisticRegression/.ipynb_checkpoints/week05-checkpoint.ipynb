```python
#Linear Regression
import numpy as np

input_data = np.array([[3,5],
                      [1,-1],
                      [0,0],
                      [8,4]])
weight = np.array([[2],[7]])
target = np.array([[33],[-5],[0],[31]])
bias = 0
```

```python
learning_rate = 0.001
epoch = 100
```

```python
def predict (X,W,b):
    return X.dot(W)+b

def get_gradient(X, error) :
    return X.T.dot(error), np.sum(error)
    
def update(W, b, gradient, learning_rate) :
    W = W - learning_rate * gradient[0]
    b = b- learning_rate * gradient[1]
    return W, b

def cost(error) :
    return np.mean(error**2)
```

```python
for i in range(epoch) :
    pred = predict(input_data, weight, bias)
    error = pred - target
    print(f"cost at epoch {i} : {cost(error):.4f}")
    gradient = get_gradient(input_data, error)
    weight, bias = update(weight, bias , gradient, learning_rate)
    print('bias : ',bias)

print("final prediction \n",predict(input_data,weight, bias))
```
