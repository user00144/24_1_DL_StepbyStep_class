```python
import numpy as np

def sin(x, T=100) :
    return np.sin(2.0 * np.pi * x/T)

def toy_data(T=100, ampl=0.05) :
    x = np.arange(0,2*T+1)
    noise = ampl* np.random.uniform(low= -1.0, high = 1.0, size = len(x))
    return sin(x, T) + noise
```

```python
import matplotlib.pyplot as plt

T = 100
f = toy_data(T)

plt.plot(f)
plt.show()
```

```python
f_t = f[:-1]
f_t_plus_1 = f[1:]

print(len(f_t), len(f_t_plus_1))

plt.plot(f_t, f_t_plus_1, 'bo')
plt.show()
```

```python
print(len(f))

f_train = f[:160]
f_test = f[160:]
```

```python
from statsmodels.tsa.ar_model import AutoReg
from sklearn.metrics import mean_squared_error

model = AutoReg(f_train, 13, old_names=False).fit()
print('Coefficients : %s' % model.params)

predictions = model.predict(start=len(f_train), end=len(f_train)+len(f_test)-1, dynamic=False)
error = mean_squared_error(f_test, predictions)
print('Test MSE: %.3f' % error)
# plot results
plt.plot(f_test)

plt.plot(predictions, color='red')
plt.show()
```

```python
#RNN preprocessing
X_ar_train = []
y_ar_train = []

p = 13

for i in range(len(f_train)-p) :
    X_ar_train.append(f_train[i:i+p])
    y_ar_train.append(f_train[i+p])

X_ar_train = np.asarray(X_ar_train)
y_ar_train = np.asarray(y_ar_train)
X_ar_train.shape
```

```python
X_rnn_train = X_ar_train.reshape(X_ar_train.shape[0], X_ar_train.shape[1], 1)
Y_rnn_train = y_ar_train.reshape(len(y_ar_train), 1)
print(X_rnn_train.shape, Y_rnn_train.shape)
# data / lag / input_vector
```

```python
#from tensorflow.keras.layers import SimpleRNN
from tensorflow.keras.layers import LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
#from tensorflow.keras.optimizers import SGD
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.optimizers import Adam
```

```python
maxlen = p  #sequence 
n_in = 1 #   ,          .
n_hidden = 20 # n_in     
#n_hidden = p # 
n_out = 1 #   

model = Sequential()
#model.add(SimpleRNN(n_hidden,
model.add(LSTM(n_hidden, input_shape=(maxlen, n_in)))
model.add(Dense(n_out, activation='linear'))
#model.add(Activation('linear'))
model.summary()
```

```python
#model.compile(optimizer='rmsprop', loss='mean_squared_error', metrics=['mse'])
rmsprop = RMSprop(learning_rate=0.0002) 
model.compile(optimizer=rmsprop, loss='mean_squared_error', metrics=['mse'])

#  
history = model.fit(X_rnn_train, Y_rnn_train, 
                    epochs=3000,
                    verbose=0,
                    batch_size=147)
```

```python
history_dict = history.history
loss = history_dict['loss']
#val_loss = history_dict['val_loss']

epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, 'bo', label='Training loss')
#plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
X_tmp = f_train[-p:] #f_train  13 
predicted = []
for i in range(len(f_test)):
    y_tmp = model.predict(X_tmp.reshape(1, p, 1)) #  
    predicted.append(y_tmp[0][0]) #  
    X_tmp = np.append(X_tmp[1:], y_tmp) # X  
print(predicted)
```

```python
error = mean_squared_error(f_test, predicted)
print('Test MSE: %.3f' % error)

plt.plot(f_test)
plt.plot(predicted, color='red')
plt.show()
```
