```python
import csv
import numpy as np
import matplotlib.pyplot as plt

date = []
y = []
with open('daily-min-temperatures.csv', encoding='utf-8') as csvfile:
    csvreader = csv.reader(csvfile)
    next(csvreader) #skip first line
    for row in csvreader:
        #print(row)
        if row:
            date.append(row[0])
            y.append(float(row[1]))

y = np.asarray(y)
plt.plot(y)
plt.show()
```

```python
len(y)
```

```python
# lag plot 
f_t = y[:-1]
f_t_plus_1 = y[1:]

print(len(f_t), len(f_t_plus_1))

plt.plot(f_t, f_t_plus_1, 'bo')
plt.show()
```

```python
# train test set 
f_train = y[:3400]
f_test = y[3400:]

print("Train size: %d, test size: %d" % (len(f_train), len(f_test)))
```

```python
from statsmodels.tsa.ar_model import AutoReg
from sklearn.metrics import mean_squared_error
```

```python
model = AutoReg(f_train, 100, old_names=False).fit()
print('Coefficients: %s' % model.params)
# make predictions
predictions = model.predict(start=len(f_train), end=len(f_train)+len(f_test)-1, dynamic=False)
#for i in range(len(predictions)):
#    print('predicted=%f, expected=%f' % (predictions[i], f_test[i]))
error = mean_squared_error(f_test, predictions)
print('Test MSE: %.3f' % error)
# plot results
plt.plot(f_test)
plt.plot(predictions, color='red')
plt.show()
```

```python
X_ar_train = []
y_ar_train = []
p = 100
for i in range(len(f_train)-p):
    X_ar_train.append(f_train[i:i+p])
    y_ar_train.append(f_train[i+p])
#print(X_ar[-1], y_ar[-1], f_train[-11:-1])
X_ar_train = np.asarray(X_ar_train)
y_ar_train = np.asarray(y_ar_train)
X_ar_train.shape
```

```python
X_rnn_train = X_ar_train.reshape(X_ar_train.shape[0], X_ar_train.shape[1], 1)
Y_rnn_train = y_ar_train.reshape(len(y_ar_train), 1)
print(X_rnn_train.shape, Y_rnn_train.shape)

X_rnn_train, X_rnn_val = X_rnn_train[:3000], X_rnn_train[3000:]
Y_rnn_train, Y_rnn_val = Y_rnn_train[:3000], Y_rnn_train[3000:]
print(X_rnn_train.shape, Y_rnn_train.shape, X_rnn_val.shape, Y_rnn_val.shape)
```

```python
#from tensorflow.keras.layers import SimpleRNN
from tensorflow.keras.layers import LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
#from tensorflow.keras.optimizers import SGD
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.optimizers import Adam
```

```python
maxlen = p  #sequence 
n_in = 1 #  -   
n_hidden = 64 # hidden node 
n_out = 1 #    

model = Sequential()
model.add(LSTM(n_hidden, input_shape=(maxlen, n_in)))
model.add(Dense(n_out, activation='linear'))
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999), loss='mean_squared_error', metrics=['mse'])
#  
history = model.fit(X_rnn_train, Y_rnn_train, 
                    epochs=100,
                    validation_data=(X_rnn_val, Y_rnn_val),
                    verbose=0,
                    batch_size=64)
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(loss) + 1)

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

```python
X_tmp = f_train[-p:]
predicted = []
for i in range(len(f_test)):
    y_tmp = model.predict(X_tmp.reshape(1, p, 1))
    predicted.append(y_tmp[0][0])
    X_tmp = np.append(X_tmp[1:], y_tmp[0][0])
    #print(X_tmp)
print(len(predicted))
```

```python
error = mean_squared_error(f_test, predicted)
print('Test MSE: %.3f' % error)

plt.plot(f_test)
plt.plot(predicted, color='red')
plt.show()
```
