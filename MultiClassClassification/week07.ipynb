```python
import numpy as np
import matplotlib.pyplot as plt

def sample(s_size):
    red_mean = [-1,0]  # The mean of the red class
    blue_mean = [1,0]  # The mean of the blue class
    green_mean = [0,2]  # The mean of the green class
    std_dev = 0.7  # standard deviation
    x_red = np.random.randn(s_size, 2) * std_dev + red_mean
    x_blue = np.random.randn(s_size, 2) * std_dev + blue_mean
    x_green = np.random.randn(s_size, 2) * std_dev + green_mean
    x = np.vstack((x_red, x_blue, x_green))
    y = np.vstack((np.zeros((s_size,1), dtype=np.int64), 
    np.ones((s_size,1), dtype=np.int64), 
    np.ones((s_size, 1), dtype=np.int64)+1))
    
    return (x, y)

input_data, target = sample(100)
x_red = input_data[(target == 0).flatten()]
x_blue = input_data[(target == 1).flatten()]
x_green = input_data[(target == 2).flatten()]
plt.plot(x_red[:,0], x_red[:,1], 'ro', label='class red')
plt.plot(x_blue[:,0], x_blue[:,1], 'b^', label='class blue')
plt.plot(x_green[:,0], x_green[:,1], 'gs', label='class green')
plt.grid()
plt.legend(loc=2)
plt.xlabel('$x_1$', fontsize=15)
plt.ylabel('$x_2$', fontsize=15)
plt.title('red vs. blue vs. green classes in the input space')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input_data, 
target, test_size=0.2, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, 
y_test.shape)
```
