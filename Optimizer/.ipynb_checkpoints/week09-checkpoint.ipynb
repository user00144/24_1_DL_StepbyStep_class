```python
import numpy as np
import matplotlib.pyplot as plt

def sample(s_size):
    red_mean = [-1,0]  # The mean of the red class
    blue_mean = [1,0]  # The mean of the blue class
    green_mean = [0,2]  # The mean of the green class
    std_dev = 0.7  # standard deviation
    x_red = np.random.randn(s_size, 2) * std_dev + red_mean
    x_blue = np.random.randn(s_size, 2) * std_dev + blue_mean
    x_green = np.random.randn(s_size, 2) * std_dev + green_mean
    x = np.vstack((x_red, x_blue, x_green))
    y = np.vstack((np.zeros((s_size,1), dtype=np.int64), 
    np.ones((s_size,1), dtype=np.int64), 
    np.ones((s_size, 1), dtype=np.int64)+1))
    
    return (x, y)

input_data, target = sample(100)
x_red = input_data[(target == 0).flatten()]
x_blue = input_data[(target == 1).flatten()]
x_green = input_data[(target == 2).flatten()]
plt.plot(x_red[:,0], x_red[:,1], 'ro', label='class red')
plt.plot(x_blue[:,0], x_blue[:,1], 'b^', label='class blue')
plt.plot(x_green[:,0], x_green[:,1], 'gs', label='class green')
plt.grid()
plt.legend(loc=2)
plt.xlabel('$x_1$', fontsize=15)
plt.ylabel('$x_2$', fontsize=15)
plt.title('red vs. blue vs. green classes in the input space')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input_data, 
target, test_size=0.2, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, 
y_test.shape)
```

```python
#One-hot encoding
class_num= len(set(y_train.flatten()))
Y_train= np.eye(class_num)[y_train.flatten()]
```

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=1, keepdims=True)
def cost(yp, y):
    return -np.multiply(y, np.log(yp)).mean()
def gradient(x, error):
    return x.T.dot(error), np.sum(error, axis=0, keepdims=True)
def predict(x, w, b):
    return softmax(x.dot(w)+b)
def update(w, b, gd, learning_rate):
    return w-learning_rate*gd[0], b-learning_rate*gd[1]
def update_momentum(w, b, gd, learning_rate, mt, v) :
    v = mt*v + learning_rate*gd[0]
    return v, w - v, b-learning_rate*gd[1]

mt = 0.9
v = np.zeros((X_train.shape[1], Y_train.shape[1]), dtype='float32')


#batch



 #초기화
weight = np.zeros((X_train.shape[1], Y_train.shape[1]))
bias = np.zeros((1, Y_train.shape[1]))
train_costs = []
```

```python
learning_rate = 0.000001
epoch = 1000
epoch_cost = 0
batch_size = 64 #batch size
batch_num = X_train.shape[0] #number of batch

for i in range(epoch):
    for j in range(batch_num) :
        X_batch = X_train[j*batch_size:(j+1)*batch_size]
        Y_batch = y_train[j*batch_size:(j+1)*batch_size]
        pred = predict(X_batch, weight, bias)
        epoch_cost += cost(pred, Y_batch)
        gd = gradient(X_batch, pred-Y_batch)
        v, weight, bias = update_momentum(weight, bias, gd, learning_rate, mt, v)
    epoch_cost = epoch_cost / batch_size
    train_costs.append(epoch_cost)

# for i in range(epoch):
#     pred = predict(X_train, weight, bias)
#     epoch_cost = cost(pred, Y_train)
#     train_costs.append(epoch_cost)
#    print(f"cost at epoch {i+1}: {epoch_cost:.4f}")
    
# Momentum : "v"
# Nesterov accelerated Gradient : get momentum -> weight update -> predict -> error -> get gradient
# Adaptive Gradient : G = G + gradient^2 -> 1/sqrt( G + e )
# RMSProp : G = rG + (1-r)gradient^2
# Momentum + RMSProp : Adam Optimizer -> weighted mean
    
    # gd = gradient(X_train, pred - Y_train)
    # v, weight, bias = update_momentum(weight, bias, gd, learning_rate, mt, v)
plt.plot(np.arange(len(train_costs)), train_costs)
plt.show()
```

```python
def acc_score(pred, target):
    return np.sum(pred==target.flatten())/len(target.flatten())
train_pred = np.argmax(predict(X_train, weight, bias), 
axis=1)
test_pred = np.argmax(predict(X_test, weight, bias), axis=1)
print(f'Train accuracy: {acc_score(train_pred, y_train):.3f}')
print(f'Test accuracy: {acc_score(test_pred, y_test):.3f}')
```
